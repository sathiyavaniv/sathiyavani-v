```python
df["TotalCharges"] = df["TotalCharges"].str.strip()   # remove spaces
```

```python
df["TotalCharges"] = df["TotalCharges"].replace("", np.nan)  # blank → NaN
```

```python
      # convert to number
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(df["MonthlyCharges"])  # fill missing
```

```python
print(f"  Fixed! Missing values now: {df['TotalCharges'].isnull().sum()}\n")
```

```python
# ── STEP 4: FIX SERVICE COLUMNS ───────────
# Problem: Some rows say "No internet service" or "No phone service"
# We simplify these to just 0 (meaning: No)
print("STEP 4: Cleaning service columns...")

service_columns = ["OnlineSecurity", "OnlineBackup", "DeviceProtection",
                   "TechSupport", "StreamingTV", "StreamingMovies", "MultipleLines"]

for col in service_columns:
    df[col] = df[col].replace("No internet service", "No")
    df[col] = df[col].replace("No phone service", "No")
    df[col] = df[col].map({"Yes": 1, "No": 0})
    print(f"  {col} → cleaned")

print()
```

```python
# ── STEP 5: CONVERT gender ─────────────────
print("STEP 5: Converting gender to 1 and 0...")
df["gender"] = df["gender"].map({"Male": 1, "Female": 0})
print("  Male = 1, Female = 0\n")
```

```python
# ── STEP 6: CONVERT SeniorCitizen ─────────
# It's already 0 and 1 — just rename for clarity
print("STEP 6: SeniorCitizen is already 0/1 — no change needed")
print("  0 = Not a senior citizen")
print("  1 = Senior citizen\n")
```

```python
# ── STEP 11: ADD 2 SIMPLE NEW COLUMNS ─────
print("STEP 11: Adding 2 new useful columns...")

# How many add-on services does each customer have?
addon_cols = ["OnlineSecurity", "OnlineBackup", "DeviceProtection",
              "TechSupport", "StreamingTV", "StreamingMovies"]
df["NumAddons"] = df[addon_cols].sum(axis=1)
print("  NumAddons = total number of extra services (0 to 6)")
```

```python
df.head()
```

```python
# Group customers by how long they have been with the company
def tenure_group(months):
    if months <= 12:
        return "New"        # 0–12 months
    elif months <= 24:
        return "Growing"    # 13–24 months
    elif months <= 48:
        return "Established"# 25–48 months
    else:
        return "Loyal"      # 49+ months

df["CustomerGroup"] = df["tenure"].apply(tenure_group)
print("  CustomerGroup = New / Growing / Established / Loyal\n")
```

```python
df.to_csv("churn_data.csv", index=False)
```

```python
import pandas as pd
import os
from sqlalchemy import create_engine

# load your cleaned file
df = pd.read_csv("churn_data.csv")

# connect to MySQL — credentials are read from environment variables,
# never hardcoded, so this notebook is safe to share/upload publicly
MYSQL_USER = os.environ.get("MYSQL_USER", "root")
MYSQL_PASSWORD = os.environ.get("MYSQL_PASSWORD")
MYSQL_HOST = os.environ.get("MYSQL_HOST", "localhost")
MYSQL_DB = os.environ.get("MYSQL_DB", "churn_db")

engine = create_engine(
    f"mysql+mysqlconnector://{MYSQL_USER}:{MYSQL_PASSWORD}@{MYSQL_HOST}:3306/{MYSQL_DB}"
)

# upload to MySQL
df.to_sql(name='customer_churn_data', con=engine, if_exists='replace', index=False)

print("Data loaded into MySQL successfully")
```
